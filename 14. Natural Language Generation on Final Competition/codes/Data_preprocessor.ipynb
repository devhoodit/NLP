```python
import pandas as pd
import json
```

```python
file_path = "drive/MyDrive/01. 강의/2023.03 NLP/14. Natural Language Generation on Final Competition/codes/data/daily_life_val.json"
```

```python
with open(file_path, 'r') as file:
    data = json.load(file)
```

```python
data = data["data"]
```

```python
df = pd.DataFrame(columns=["dialogueID", "utteranceID","turnID","participantID","date","time","utterance"])
```

```python
for idx in range(len(data)):
  dialogueID = data[idx]["header"]["dialogueInfo"]["dialogueID"]
  numberOfParticipants = data[idx]["header"]["dialogueInfo"]["numberOfParticipants"]
  topic = data[idx]["header"]["dialogueInfo"]["topic"]
  
  for idx_att in range(len(data[idx]["body"])):
    an_att = data[idx]["body"][idx_att]
    an_att["dialogueID"] = dialogueID
    a_df = pd.DataFrame([an_att])
    df = pd.concat([df,a_df])
```
